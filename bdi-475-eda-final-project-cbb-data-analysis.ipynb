```python
#filter dataframe for 2021 to use most recent CBB data

df_cbb = pd.read_csv("/kaggle/input/college-basketball-dataset/cbb.csv")
df_cbb_21 = pd.read_csv("/kaggle/input/college-basketball-dataset/cbb21.csv")
```

```python
#I wanted to see how big of an effect offensive efficiency had on number of Wins

fig = px.scatter(df_cbb_21, x='ADJOE', y='W', trendline='ols', title='Wins by Offensive Efficiency in 2021')
fig.show()
```

```python
#Conversely, I also wanted to see how big of an effect defensive efficiency had on number of wins
#I had to reverse the x-axis b/c a lower ADJDE means that they have a better defense
#Defensive efficiency matters marginally more than offensive efficiency

fig = px.scatter(df_cbb_21, x='ADJDE', y='W', trendline='ols', title='Wins by Defensive Efficiency in 2021')
fig.update_xaxes(autorange="reversed")
fig.show()
```

```python
#I decided to filter into a dataframe with only the 'Power 6' conferences with the vast majority of major basketball programs

df_p6 = df_cbb_21[(df_cbb_21['CONF'] == 'B10') | (df_cbb_21['CONF'] == 'ACC') | (df_cbb_21['CONF'] == 'SEC') | (df_cbb_21['CONF'] == 'B12') | (df_cbb_21['CONF'] == 'P12') | (df_cbb_21['CONF'] == 'BE')]
```

```python
#I wanted to find the averages by Power 6 conferences for all statistics

df_p6_by_conf = df_p6.groupby(
    df_p6['CONF'],
    as_index=False
).mean()

display(df_p6_by_conf)
```

```python
#I wanted to vizualize which P6 conferences had the best teams on average
#The highest bars should theoretically be the highest quality conferences for 2021

fig = px.bar(df_p6_by_conf, x='CONF', y='BARTHAG', title='Average Power Rating by Power 6 Conference')
fig.show()
```

```python
#I wanted to visualize what the fastest conferences were (on average)
#Because the disparities are quite small, I had to update the y axis range to zoom in

fig = px.bar(df_p6_by_conf, x='CONF', y='ADJ_T', title='Average Tempo by Conference')
fig.update_yaxes(range=[67, 70])
fig.show()
```

```python
#I wanted to determine whether Tempo of play impacted quality of team (Power ranking)

fig = px.scatter(df_cbb_21, x='ADJ_T', y='BARTHAG', trendline='ols', title='Power Rating by Tempo in 2021')
fig.show()
```

```python
#I wanted to find how intense the positive relationship was between Power Rating & Seed
#Theoretically, each bar should be increasing from left to right, or else that means that the selection committee got their picks wrong


fig = px.bar(df_cbb_21, x='SEED', y='BARTHAG', barmode='overlay', title='March Madness Seed by Power Rating')
fig.update_xaxes(autorange="reversed")
fig.show()
```
